```python
#WEEK 3
# === Feature Engineering and Selection ===

# Correlation threshold method
corr_matrix = df[numerical_features].corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
print("Highly correlated features to drop:", to_drop)

df_reduced = df.drop(columns=to_drop)

# Feature importance using Random Forest
from sklearn.ensemble import RandomForestClassifier
X = df_reduced.drop('Label', axis=1)
y = df_reduced['Label']

rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X, y)

importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(10,6))
importances[:15].plot(kind='bar')
plt.title("Top 15 Important Features")
plt.show()
```

```python
# === WEEK 4: Supervised Model Training ===

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# --- Data Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- Handle Missing Values ---
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# --- Feature Scaling ---
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# --- Random Forest Classifier ---
print("🔹 Training Random Forest...")
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print("\nRandom Forest Report:\n", classification_report(y_test, y_pred_rf))

# --- Support Vector Machine (SVM) ---
# ⚡ Tip: SVM can be slow for large datasets. Use a subset if needed.
print("🔹 Training SVM (RBF Kernel)...")
if len(X_train) > 20000:  # if dataset is huge, train on subset
    X_train_svm = X_train[:20000]
    y_train_svm = y_train[:20000]
else:
    X_train_svm = X_train
    y_train_svm = y_train

svm = SVC(kernel='rbf', gamma='scale', C=1.0)
svm.fit(X_train_svm, y_train_svm)
y_pred_svm = svm.predict(X_test)
print("\nSVM Report:\n", classification_report(y_test, y_pred_svm))

# --- Logistic Regression ---
print("🔹 Training Logistic Regression...")
lr = LogisticRegression(max_iter=500, solver='lbfgs', n_jobs=-1)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
print("\nLogistic Regression Report:\n", classification_report(y_test, y_pred_lr))

print("\n✅ Training Completed Successfully!")
```

```python
# === WEEK 5: Anomaly Detection ===

import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

# --- Handle Missing Values ---
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# --- Feature Scaling (important for KMeans and IsolationForest) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# --- Isolation Forest for Anomaly Detection ---
print("🔹 Running Isolation Forest...")
iso_forest = IsolationForest(contamination=0.02, random_state=42, n_jobs=-1)
anomalies = iso_forest.fit_predict(X_scaled)

# Add results to DataFrame
df['Anomaly_IF'] = anomalies
print("\nIsolation Forest Anomaly Counts:")
print(df['Anomaly_IF'].value_counts())

# Replace -1 (anomalies) and 1 (normal) with labels for readability
df['Anomaly_IF_Label'] = df['Anomaly_IF'].map({1: 'Normal', -1: 'Anomaly'})

# --- K-Means Clustering ---
print("\n🔹 Running K-Means Clustering...")
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Add results to DataFrame
df['Cluster'] = clusters

# --- Visualization ---
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=X_scaled[:, 0],
    y=X_scaled[:, 1],
    hue=df['Cluster'],
    palette='viridis',
    s=10
)
plt.title("K-Means Clustering on Network Data")
plt.xlabel("Feature 1 (scaled)")
plt.ylabel("Feature 2 (scaled)")
plt.show()

# --- Optional: Show anomaly percentage ---
total = len(df)
anomalies_count = (df['Anomaly_IF'] == -1).sum()
print(f"\n⚠️ Detected {anomalies_count} anomalies out of {total} samples "
      f"({(anomalies_count / total * 100):.2f}%).")

print("\n✅ Anomaly Detection Completed Successfully!")
```

```python
# === WEEK 6: Model Evaluation and Fine-Tuning (Optimized for Colab) ===

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    roc_curve, roc_auc_score, classification_report,
    confusion_matrix, roc_auc_score
)
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, label_binarize
import joblib
from itertools import cycle

# --- Step 1: Split Data ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# --- Step 2: Handle Missing Values ---
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# --- Step 3: Feature Scaling ---
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# --- Step 4: Random Forest Grid Search ---
print("🔹 Starting Grid Search for Random Forest...")

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}

grid = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42, n_jobs=-1, class_weight='balanced'),
    param_grid=param_grid,
    cv=3,
    scoring='accuracy',
    verbose=1,
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("\n✅ Grid Search Completed Successfully!")
print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation Score:", round(grid.best_score_, 4))

# --- Step 5: Evaluate on Test Set ---
best_rf = grid.best_estimator_
y_pred = best_rf.predict(X_test)

print("\nClassification Report (Test Set):\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# --- Step 6: ROC Curve (Supports Binary + Multiclass) ---
if hasattr(best_rf, "predict_proba"):
    y_prob = best_rf.predict_proba(X_test)
    n_classes = y_prob.shape[1]

    if n_classes == 2:
        # Binary ROC
        fpr, tpr, _ = roc_curve(y_test, y_prob[:, 1])
        auc_score = roc_auc_score(y_test, y_prob[:, 1])
        plt.figure(figsize=(6, 5))
        plt.plot(fpr, tpr, label=f"AUC = {auc_score:.3f}", linewidth=2)
        plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title("ROC Curve - Random Forest (Binary)")
        plt.legend()
        plt.grid(True)
        plt.show()

    else:
        # Multiclass ROC (One-vs-Rest)
        classes = np.unique(y_test)
        y_test_bin = label_binarize(y_test, classes=classes)
        auc_dict = {}

        plt.figure(figsize=(7, 6))
        colors = cycle(["blue", "red", "green", "orange", "purple", "brown"])

        for i, color in zip(range(n_classes), colors):
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
            auc = roc_auc_score(y_test_bin[:, i], y_prob[:, i])
            auc_dict[classes[i]] = auc
            plt.plot(fpr, tpr, color=color, lw=2,
                     label=f"Class {classes[i]} (AUC = {auc:.3f})")

        plt.plot([0, 1], [0, 1], "k--", lw=1)
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title("ROC Curve - Random Forest (Multiclass)")
        plt.legend(loc="lower right")
        plt.grid(True)
        plt.show()

        print("\nClass-wise AUC Scores:")
        for cls, auc_val in auc_dict.items():
            print(f"Class {cls}: AUC = {auc_val:.3f}")
else:
    print("⚠️ This model does not support probability estimates.")

# --- Step 7: Save the Best Model for Future Use ---
joblib.dump(best_rf, "best_random_forest_model.pkl")
print("\n💾 Model saved as best_random_forest_model.pkl")
print("\n✅ Model Fine-Tuning and Evaluation Completed Successfully!")
```

```python
# ============================================================
# WEEK 7 – FINAL AUTO VERSION (Uses df and X from above cells)
# ============================================================

import pandas as pd
import numpy as np
import datetime
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

# =====================================================
# 1️⃣ CHECK IF df and X EXIST
# =====================================================
if 'df' not in globals() or 'X' not in globals():
    raise RuntimeError("❌ ERROR: df and X must already exist. Run previous cells first.")

# =====================================================
# 2️⃣ CHECK LABEL COLUMN
# =====================================================
if "Label" not in df.columns:
    raise KeyError("❌ ERROR: Dataset must contain a 'Label' column.")

# =====================================================
# 3️⃣ LOAD OR TRAIN MODEL
# =====================================================
MODEL_FILE = "random_forest_model.pkl"

try:
    rf = joblib.load(MODEL_FILE)
    print("✅ Loaded saved RandomForest model.")
except:
    print("⚠️ No saved model found — training a new RandomForest model...")

    X_train, X_test, y_train, y_test = train_test_split(
        X, df["Label"], test_size=0.2, random_state=42
    )

    rf = RandomForestClassifier(n_estimators=200, random_state=42)
    rf.fit(X_train, y_train)

    joblib.dump(rf, MODEL_FILE)
    print("✅ Model trained and saved.")

# =====================================================
# 4️⃣ HANDLE MISSING VALUES
# =====================================================
imputer = SimpleImputer(strategy="mean")
X_imputed = imputer.fit_transform(X)

# =====================================================
# 5️⃣ GENERATE PREDICTIONS
# =====================================================
print("🔹 Generating predictions using RandomForest...")
df["Prediction_RF"] = rf.predict(X_imputed)

# =====================================================
# 6️⃣ IDENTIFY ALERTS (mismatches)
# =====================================================
alerts = df[df["Prediction_RF"] != df["Label"]].copy()
alerts["Timestamp"] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Safe column filtering
key_cols = [
    col for col in ["Timestamp", "Flow Duration", "Label", "Prediction_RF"]
    if col in alerts.columns
]

alerts_to_save = alerts[key_cols]

# =====================================================
# 7️⃣ SAVE ALERT LOG
# =====================================================
output_file = "alerts_log.csv"
alerts_to_save.to_csv(output_file, index=False)

# =====================================================
# 8️⃣ FINAL OUTPUT
# =====================================================
print(f"\n✅ Total Alerts Generated: {len(alerts)}")
print(f"📁 Alerts saved to: {output_file}")

if not alerts.empty:
    print("\n🔍 Preview of alerts:")
    print(alerts_to_save.head())
else:
    print("\n✅ No mismatches — No alerts generated.")
```
